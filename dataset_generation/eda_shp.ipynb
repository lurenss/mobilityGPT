```python
import pandas as pd
import folium

# Function to parse path data into a list of coordinates
def parse_path(path_str):
    path = eval(path_str)
    return [(lat, lon) for lon, lat, _ in path]

# Load the dataset
file_path = './gps_trace.csv'  # Replace with your file path
gps_data = pd.read_csv(file_path)

# Apply the function to parse paths
gps_data['parsed_path'] = gps_data['path'].apply(parse_path)

# Calculate average latitude and longitude
latitudes = []
longitudes = []

for path in gps_data['parsed_path']:
    for lat, lon in path:
        latitudes.append(lat)
        longitudes.append(lon)

avg_latitude = sum(latitudes) / len(latitudes)
avg_longitude = sum(longitudes) / len(longitudes)

# Create a base map
m = folium.Map(location=[avg_latitude, avg_longitude], zoom_start=12)

# Define colors for each user
user_colors = {user: f'#{user:06x}' for user in range(len(gps_data['User'].unique()))}

# Plotting each path as a series of dots on the map
for user, group in gps_data.groupby('User'):
    path_color = user_colors.get(user, 'black')  # Use black color if user is not in the dictionary
    for path in group['parsed_path']:
        for lat, lon in path:
            folium.CircleMarker([lat, lon], radius=3, color=path_color, fill=True).add_to(m)

# display the map
m

```
